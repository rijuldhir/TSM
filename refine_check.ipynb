```python
import json
import csv
with open('check.txt') as f:
    read = f.readlines()
```

```python
count =0
lis = []
for i in range(1,len(read[1:]),5):
    x = read[i:i+5]
    pq = []
    for y in x:
        y = y.strip().replace(']','')
        y = y.replace('[',' ')
        if len(y) > 4:
            pq.extend(y.split(' '))
        else:
            pq.append(y.strip())
    pq = [i for i in pq if i]
        
    lis.append(pq)
print(lis)
```

```python
with open('tools/category.txt') as f:
    read = f.readlines()
#print(read)
rev_dic = {}
count = 0
for b in read:
    rev_dic[str(count)] = b.strip()
    count += 1

label_list = []

for x in lis:
    pq = []
    for y in x:
        pq.append(rev_dic[y].split(' ')[0])
    label_list.append(pq)
print(len(label_list))
```

```python
print(label_list[0])
```

```python
print('Top 1 verb accuracy for pretrained one :-')
correct = 0
count = 0
for x in label_list:
    if x[0] == 'Something' or x[0] == 'Pretending':
        continue
    count += 1
    if x[0] == x[1]:
        correct += 1
print(correct*100/count)
```

```python
print('Top 5 verb accuracy for pretrained one :-')
correct = 0
count  = 0
for x in label_list:
    if x[0] == 'Something' or x[0] == 'Pretending':
        continue
    count += 1
    if x[0] == x[1] or x[0] == x[2] or x[0] == x[3] or x[0] == x[4] or x[0] == x[5]:
        correct += 1
print(correct*100/count)
```
